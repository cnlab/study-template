```python
%matplotlib inline
from IPython.display import HTML

from bs4 import BeautifulSoup, Tag

import matplotlib.pyplot as plt
from nilearn import plotting,image
import nilearn

import re
import os
import pandas as pd
```

### Define variables
* `pID` = participant ID including 'sub-'
* `relative_folder` = path to the fmriprep folder relative to this file

```python
pID='sub-BPP01'
relative_folder = '../../data/bids_data/derivatives/fmriprep/'
```

### Compile report file

```python
html = open('{}/{}.html'.format(relative_folder, pID)).read()
dom = BeautifulSoup(html,'html.parser')
for obj in dom.find_all(href=re.compile("^./")):
    src=re.sub('^./',relative_folder,obj['href'])
    img_tag = BeautifulSoup('<embed src="{}"/>'.format(src), 'html.parser')
    obj.replaceWith(img_tag)
    
HTML(str(dom))
```
